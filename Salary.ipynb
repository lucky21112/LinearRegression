```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Salary.csv.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['Salary'])
```

```python
sns.distplot(df['YearsExperience'])
```

```python
# Checking normality assumption
plt.scatter(df['YearsExperience'],df['Salary'])
```

```python
# linearly distribution assumption true
```

```python
target=df['Salary']
input=df['YearsExperience']
input=input.values.reshape(-1,1)
reg=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.3,random_state=365)
```

```python
reg.fit(x_train,y_train)
y_hat=reg.predict(x_train)
```

```python
sns.distplot(y_train-y_hat)
```

```python
# Mean=0 of residual plot normaly distrbuted all assumption fulfield
```

```python
reg.score(x_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_hat)
```

```python
# How accurate it is preedicting
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_train,y_hat)
```
