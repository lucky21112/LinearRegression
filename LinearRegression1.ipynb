```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sms
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df=pd.read_csv('Added_New_Variable.csv')
```

```python
df.head()
```

```python
data_raw=df.copy()
df['Attendance']=df['Attendance'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
y=df['GPA']
x1=df[['SAT','Attendance']]
x=sms.add_constant(x1)
results=sms.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(df['SAT'],y)
yhat_no=0.6439 +0.0014*df['SAT']
yhat_yes=0.8605+0.0014*df['SAT']
fig=plt.plot(df['SAT'],yhat_no,color='blue')
plt.plot(df['SAT'],yhat_yes,color='yellow')
```
